## Step 3: Transforming data 

```python
data['Date']=pd.to_datetime(data['Date'],format='%Y-%m-%d')
data['Date']
```

```python
#show data types of the converted data
data.dtypes
```

```python
#Extract the year, month, day attributes of the datetime objects converted for the Date feature
#create the corresponding new features Year, Month and Day

data['Year']=data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data
```

```python
data = data.drop(['Date'],axis=1)
data
```

## Step 4: Preparing Training and Testing Datasets
Prepare data - feature set and target set

Build target set Y by extracting 'RainTomorrow' from data
Build feature set X by excluding/dropping the 'RainTomorrow' from data

```python
Y=data['RainTomorrow']
X=data.drop(['RainTomorrow'], axis=1)
print(X)
print(Y)
```

### Step 4.2
Split the data into training and testing datasets

Using the sklearn.model_selection.train_test_split() function to split the training data, feature set and target set (i.e. X and Y), into two subsets for training and testing, i.e. X_train, X_test, Y_train, Y_test.

Set test_size to 0.3. test_size is the proportion of the dataset to include in the test split.
Set random_state to 0. This integer value is the seed used by the random number generator.
Set stratify to stratify the dataset based on the target labels (i.e. Y).

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0, stratify=Y)
```

```python
#show info of the splited sets
print("X_train : " + str(X_train.shape))
print("X_test : " + str(X_test.shape))
print("y_train : " + str(Y_train.shape))
print("y_test : " + str(Y_test.shape))
```

## Step 5: Cleaning data: Handling missing values
### Step 5.1 Missing values in the training dataset
count how many values were missing in the dataset

```python
X_train.isnull().sum()
```

### Step 5.2 Categorical features with missing values of the training dataset
use the most common value of the feature to impute the missing values. 

```python
X_train["WindGustDir"].fillna(X_train['WindGustDir'].mode().iloc[0], inplace=True)
X_train["WindDir9am"].fillna(X_train['WindDir9am'].mode().iloc[0], inplace=True)
X_train["WindDir3pm"].fillna(X_train['WindDir3pm'].mode().iloc[0], inplace=True)
X_train["RainToday"].fillna(X_train["RainToday"].mode().iloc[0], inplace=True)
```

### Step 5.3 Numerical features with missing values of the training dataset
Use mean / median to impute the missing values of the feature

```python
numeric_list = ["MinTemp", "MaxTemp", "Rainfall", "Evaporation", "Sunshine", "WindGustSpeed", "WindSpeed9am", "WindSpeed3pm", "Humidity9am", "Humidity3pm",
"Pressure9am", "Pressure3pm", "Cloud9am", "Cloud3pm", "Temp9am", "Temp3pm"]

for i in numeric_list:
    X_train[i].fillna(X_train[i].mean(), inplace=True)
```

### Step 5.4 Handling missing values of the testing dataset 

```python
X_test["WindGustDir"].fillna(X_test['WindGustDir'].mode().iloc[0], inplace=True)
X_test["WindDir9am"].fillna(X_test['WindDir9am'].mode().iloc[0], inplace=True)
X_test["WindDir3pm"].fillna(X_test['WindDir3pm'].mode().iloc[0], inplace=True)
X_test["RainToday"].fillna(X_test["RainToday"].mode().iloc[0], inplace=True)
```

```python
numeric_list = ["MinTemp", "MaxTemp", "Rainfall", "Evaporation", "Sunshine", "WindGustSpeed", "WindSpeed9am", "WindSpeed3pm", "Humidity9am", "Humidity3pm",
"Pressure9am", "Pressure3pm", "Cloud9am", "Cloud3pm", "Temp9am", "Temp3pm"]

for i in numeric_list:
    X_test[i].fillna(X_test[i].mean(), inplace=True)
```

## Step 6 convert categorical features into dummy/indicator features of the dataset

```python
X_train = pd.get_dummies(X_train)
X_train.head(10)
```

```python
X_test = pd.get_dummies(X_test)
X_test.head(10)
```

### Step 6.1 Align the testing dataset with the training dataset 

```python
X_train, X_test = X_train.align(X_test, join="left", axis=1)
X_test.head(10)
```

```python
print("X_train : " + str(X_train.shape))
print("X_test : " + str(X_test.shape))
print("y_train : " + str(Y_train.shape))
print("y_test : " + str(Y_test.shape))
```

## Step 7: Training the Regression model using Sklearn
Build the logistic regression model by:

Initialize a Logistic Regression model by LogisticRegression() function.
Call fit() function to train the logistic regression model using the training feature data (i.e. X_train).
Predict the target values for the testing feature data (i.e. X_test) using the predict() function.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
y_pred
```

## step 7.1 Evaluation

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_test, y_pred)
```

## Step 8 - Standardization and Rescaling (10 points)
Rescaling and Standardization of datasets is a common requirement for many machine learning estimators implemented in scikit-learn. They might behave badly if the individual features do not more or less look like standard normally distributed data: Gaussian with zero mean and unit variance.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
trans_X_train = scaler.transform(X_train)
X_train = pd.DataFrame(trans_X_train, columns=X_train.columns)
X_train.head()

scaler.fit(X_test)
trans_X_test = scaler.transform(X_test)
X_test = pd.DataFrame(trans_X_test, columns=X_test.columns)
X_test.head()
```

## Step 9 -- Training the Regression model on the rescaled and standardized dataset
Re-train the regression model using the rescaled and standardized datasets

```python
from sklearn.metrics import accuracy_score

model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

accuracy_score(Y_test, y_pred)
```
